```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

load_dotenv("C:/Users/chira/.env")
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

model = genai.GenerativeModel('gemini-2.5-flash')

response = model.generate_content("Say hello. Confirm API is working.")
print("✅ API is working!")
print("\nResponse:")
print(response.text)
```

```python
import pandas as pd

df = pd.read_csv("data/category_sales_summary.csv")   # change path if needed

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
summary = f"""
Total Sales: ${df['Sales'].sum():,.2f}
Average Sales: ${df['Sales'].mean():,.2f}
Maximum Sale: ${df['Sales'].max():,.2f}
Minimum Sale: ${df['Sales'].min():,.2f}
Total Orders: {df.shape[0]:,}

Top Region by Sales:
{df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(5)}

Top Product by Sales:
{df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)}

Top Category by Sales:
{df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(5)}
"""
print(summary)
```

```python
prompt = f"""
You are a senior business intelligence analyst.
Here is the sales dataset summary:
{summary}

Provide:
1. Key trends in the business
2. Potential problems or risks
3. Strategic recommendations for management
4. Opportunities for growth

Write the analysis like an executive report.
"""

response = model.generate_content(prompt)
print("="*60)
print("📊 EXECUTIVE BUSINESS INTELLIGENCE REPORT")
print("="*60)
print(response.text)
print("="*60)
```

```python
question = "Why are sales lower in some regions?"

prompt = f"""
You are a data analyst.
Here is the dataset summary:
{summary}

Answer this question based on the data:
{question}
"""

response = model.generate_content(prompt)
print("="*60)
print(f"❓ QUESTION: {question}")
print("="*60)
print(response.text)
print("="*60)
```
