```python
import pandas as pd

df = pd.read_csv("../data/raw/sales.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
summary = f"""
Total Sales: ${df['Sales'].sum():,.2f}
Average Sales: ${df['Sales'].mean():,.2f}
Maximum Sale: ${df['Sales'].max():,.2f}
Minimum Sale: ${df['Sales'].min():,.2f}
Total Orders: {df.shape[0]:,}

Top Region by Sales:
{df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(5)}

Top Product by Sales:
{df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)}

Top Category by Sales:
{df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(5)}
"""
print(summary)
```

```python
prompt = f"""
You are a senior business intelligence analyst.
Here is the sales dataset summary:
{summary}

Provide:
1. Key trends in the business
2. Potential problems or risks
3. Strategic recommendations for management
4. Opportunities for growth

Write the analysis like an executive report.
"""

response = model.generate_content(prompt)
print("="*60)
print("ðŸ“Š EXECUTIVE BUSINESS INTELLIGENCE REPORT")
print("="*60)
print(response.text)
print("="*60)
```

```python
question = "Why are sales lower in some regions?"

prompt = f"""
You are a data analyst.
Here is the dataset summary:
{summary}

Answer this question based on the data:
{question}
"""

response = model.generate_content(prompt)
print("="*60)
print(f"â“ QUESTION: {question}")
print("="*60)
print(response.text)
print("="*60)
```

```python
import pandas as pd

# Convert Order Date to datetime (if not already done)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')

# Create Month column
df['Month'] = df['Order Date'].dt.to_period('M')

# Calculate monthly sales
monthly_sales = df.groupby('Month')['Sales'].sum().sort_index()

print("="*60)
print("ðŸ“… MONTHLY SALES TREND")
print("="*60)
print(monthly_sales)
print("="*60)

# Generate AI insight
trend_prompt = f"""
Here is the monthly sales trend:
{monthly_sales}

Analyze this data and:
1. Explain the overall trend (increasing, decreasing, stable)
2. Identify any anomalies or unusual patterns
3. Suggest possible reasons for the trends
4. Recommend actions based on the pattern
"""

response = model.generate_content(trend_prompt)
print("\n" + "="*60)
print("ðŸ¤– AI TREND ANALYSIS")
print("="*60)
print(response.text)
print("="*60)
```

```python
category_summary = df.groupby('Category').agg({
    'Sales': ['sum', 'mean', 'count'],
    'Product Name': 'nunique'
}).round(2)

prompt = f"""
You are a product analyst.
Here is the category performance data:
{category_summary}

Analyze:
1. Which category is performing best and why?
2. Which category needs attention?
3. What strategies would you recommend for each category?
"""

response = model.generate_content(prompt)
print("="*60)
print("ðŸ“¦ CATEGORY PERFORMANCE ANALYSIS")
print("="*60)
print(response.text)
```

```python
segment_summary = df.groupby('Segment').agg({
    'Sales': ['sum', 'mean'],
    'Customer ID': 'nunique'
}).round(2)

prompt = f"""
You are a customer analytics expert.
Here is the customer segment data:
{segment_summary}

Provide:
1. Which segment is most valuable?
2. How should we approach each segment differently?
3. What marketing strategies would work for each?
"""

response = model.generate_content(prompt)
print("="*60)
print("ðŸ‘¥ CUSTOMER SEGMENT INSIGHTS")
print("="*60)
print(response.text)
```
