```python
import pandas as pd

df = pd.read_csv("../data/raw/sales.csv")
category_summary = df.groupby("Category", as_index=False)["Sales"].sum()
category_summary.head()
```

```python
summary_text = category_summary.to_string(index=False)

print(summary_text)
```

```python
!pip install openai

client = OpenAI(api_key="YOUR_API_KEY_HERE")
```

```python
prompt = f"""
You are a business intelligence analyst.

Analyze the following sales data and generate 3 executive insights.

Sales Data:
{summary_text}

Focus on:
- Top performing category
- Relative performance differences
- Strategic recommendations
"""

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role": "user", "content": prompt}]
)

print(response.choices[0].message.content)
```
